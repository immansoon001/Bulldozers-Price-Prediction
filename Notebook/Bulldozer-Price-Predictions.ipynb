##build your own evaluation function


```python
#create evaluation function
from sklearn.metrics import mean_squared_log_error, mean_absolute_error, r2_score
def rmsle(y_test, y_preds):
    """
    Calculate root mean squared log error between predictions and true labels.
    """
    return np.sqrt(mean_squared_log_error(y_test, y_preds))

#create functions to evaluate model on a few different levels
def show_scores(model):
    train_preds = model.predict(X_train)
    val_preds = model.predict(X_val)
    scores = {"Training MAE": mean_absolute_error(y_train, train_preds),
              "Validation MAE": mean_absolute_error(y_val, val_preds),
              "Training RMSLE": rmsle(y_train, train_preds),
              "Validation RMSLE": rmsle(y_val, val_preds),
              "Training R^2": r2_score(y_train, train_preds),
              "Validation R^2": r2_score(y_val, val_preds)}
    return scores
```

##make predictions on test data

```python
#import the test data
df_test = pd.read_csv("../data/bluebook-for-bulldozers/Test.csv",
                      low_memory=False,parse_dates=["saledate"])

df_test.head()
```

```python
##preprocessing the data(getting the test data in the same format as our training data)

def preprocess_data(df):
    """
    Performs transformations on df and returns transformed df.
    """
    df["saleYear"] = df.saledate.dt.year
    df["saleMonth"] = df.saledate.dt.month
    df["saleDay"] = df.saledate.dt.day
    df["saleDayOfWeek"] = df.saledate.dt.dayofweek
    df["saleDayOfYear"] = df.saledate.dt.dayofyear
    
    df.drop("saledate", axis=1, inplace=True)
    
    # Fill the numeric rows with median
    for label, content in df.items():
        if pd.api.types.is_numeric_dtype(content):
            if pd.isnull(content).sum():
                # Add a binary column which tells us if the data was missing or not
                df[label+"_is_missing"] = pd.isnull(content)
                # Fill missing numeric values with median
                df[label] = content.fillna(content.median())
    
        # Filled categorical missing data and turn categories into numbers
        if not pd.api.types.is_numeric_dtype(content):
            df[label+"_is_missing"] = pd.isnull(content)
            # We add +1 to the category code because pandas encodes missing categories as -1
            df[label] = pd.Categorical(content).codes+1
    
    return df


        
```

```python
#preprocess the test data
df_test = preprocess_data(df_test)
df_test.head()
```

```python
# Align test features with training features
missing_cols = set(X_train.columns) - set(df_test.columns)
print('Missing columns count:', len(missing_cols))
print('Sample missing cols:', sorted(list(missing_cols))[:30])
# Add any missing columns with default 0 and match training column order
df_test = df_test.reindex(columns=X_train.columns, fill_value=0)
# Attempt to align dtypes to X_train where possible
for c in X_train.columns:
    if c in df_test.columns:
        try:
            df_test[c] = df_test[c].astype(X_train[c].dtype)
        except Exception:
            pass
df_test.head()
```

```python
# Convert any remaining non-numeric columns in `df_test` to numeric codes
non_numeric = [c for c in df_test.columns if not pd.api.types.is_numeric_dtype(df_test[c])]
print('Non-numeric columns before conversion:', non_numeric)
for c in non_numeric:
    df_test[c] = pd.Categorical(df_test[c]).codes + 1
print('Non-numeric columns after conversion:', [c for c in df_test.columns if not pd.api.types.is_numeric_dtype(df_test[c])])
df_test.head()
```

### Feature Importance
feature importance seeks to figure out which different attributes of the data were most important when it came to predicting the target variable 

```python
#find feature importance of our ideal model
ideal_model.feature_importances_
```

```python
#helper function for plotting feature importance
def plot_features(columns, importances, n=20):
    df = (pd.DataFrame({"features": columns, "feature_importances": importances})
          .sort_values("feature_importances", ascending=False)
          .reset_index(drop=True))
    
    #plot the top n feature importances
    fig, ax = plt.subplots()
    ax.barh(df["features"][:n], df["feature_importances"][:20])
    ax.set_ylabel("Features")
    ax.set_xlabel("Feature importance")
    ax.invert_yaxis()
```

```python
plot_features(X_train.columns, ideal_model.feature_importances_)
```
